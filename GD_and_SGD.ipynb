```python
from numpy import array



######### tweak iter and learning_rate and see the behaviour of the output ####3
iter = 1000           
learning_rate = 0.008


m = c = 0.5

# x = array([1,2,3,4,5])
# y = array([3,5,7,9,11])

# x = array([1,2,3,4])
# y = array([1,3,5,7])

x = array([1,2,3,4,5,6])
y = array([3,5,7,9,11,13]) 

N = len(x)

for i in range(iter):
  yp = m*x + c

  cost = (-1/N) * sum( (yp-y)**2 )
  m = m - learning_rate*((-2/N) * sum((y-yp)*x ))
  c = c - learning_rate * ((-2/N) * sum(y-yp))
  print("m >> {}        c >> {}        cost >>  {}  ".format(m,c,cost))
```

**Stochastic Gradient Descent** (taking mini-batches of size 3)

[Additional Resources](https://www.geeksforgeeks.org/ml-stochastic-gradient-descent-sgd/)

```python
from numpy import array
from random import randint as random_number

######### tweak iter and learning_rate and see the behaviour of the output ####3
iter = 50000  
learning_rate = 0.001


m = c = 0.5


x = array([i for i in range(30)])

y = array([2*i+1 for i in range(30)]) 
btach_size = 3

N = len(x)

ran1 = random_number(0,N)
ran2 = random_number(0,ran1)
# print(N)
for i in range(iter):
  ran1 = random_number(0,N-1)
  ran2 = ran1 - btach_size
  
  yp = m*x[ran2:ran1] + c

  cost = (-1/N) * sum((yp-y[ran2:ran1])**2 )
  m = m - learning_rate*((-2/N) * sum((y[ran2:ran1]-yp)*x[ran2:ran1] ))
  c = c - learning_rate * ((-2/N) * sum((y[ran2:ran1]-yp)))
  
  print("m >> {}        c >> {}        cost >>  {}  iter >> {} ".format(m,c,cost,i))
```

**SGD using Sci-kit learn**

```python
# Run this code multiple times and see different values as output
from sklearn.linear_model import SGDRegressor

x = array([i for i in range(30)]).reshape(-1,1)
y = array([2*i+1 for i in range(30)]).reshape(-1,1)

clf = SGDRegressor(max_iter=1000000000)
clf.fit(x ,y)

```

```python
print(clf.coef_,clf.intercept_)
```
